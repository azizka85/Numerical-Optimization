### 1. Finite Difference Method

##### Example #1.1

$f(x) = (x - 2)^2$

$f'(x) = 2(x-2)$

$f''(x) = 2 > 0, \forall x$

```python
import numpy as np
from matplotlib import pyplot as plt

def f(x: np.ndarray[np.double]) -> np.double:
    return (x[0] - 2.)**2

x = np.linspace(0, 4, 20)
y = [f([t]) for t in x]

plt.xlabel('x')
plt.grid(True)
plt.plot(x, y)
plt.title('$f(x) = (x-2)^2$')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
from modules.gradient.finite_difference import FDGradient
from modules.hessian.finite_difference import FDHessian

def df(x: np.ndarray[np.double]) -> np.ndarray[np.double]:
    return np.array([2*(x[0] - 2)])

def d2f(x: np.ndarray[np.double]) -> np.ndarray[np.ndarray[np.double]]:
    return [[2]]

fd_df = FDGradient(f, 10**-3)

error = [abs(df([t])[0] - fd_df([t])[0]) for t in x]

plt.xlabel('x')
plt.grid(True)
plt.plot(x, error)
plt.title('Error of fd df(x) with $\\epsilon = 10^{-3}$')
plt.xlabel("x")
plt.show()
```

```python
fd_df = FDGradient(f, 10**-6)

error = [abs(df([t])[0] - fd_df([t])[0]) for t in x]

plt.xlabel('x')
plt.grid(True)
plt.plot(x, error)
plt.title('Error of fd df(x) with $\\epsilon = 10^{-6}$')
plt.xlabel("x")
plt.show()
```

```python
fd_d2f = FDHessian(f, 10**-3)

error = [abs(d2f([t])[0][0] - fd_d2f([t])[0][0]) for t in x]

plt.xlabel('x')
plt.grid(True)
plt.plot(x, error)
plt.title('Error of fd d2f(x) with $\\epsilon = 10^{-3}$')
plt.xlabel("x")
plt.show()
```

```python
fd_d2f = FDHessian(f, 10**-6)

error = [abs(d2f([t])[0][0] - fd_d2f([t])[0][0]) for t in x]

plt.xlabel('x')
plt.grid(True)
plt.plot(x, error)
plt.title('Error of fd d2f(x) with $\\epsilon = 10^{-6}$')
plt.xlabel("x")
plt.show()
```

```python
from modules.optimize.unconstrained.line_search.newton import newton

fd_df = FDGradient(f, 10**-3)
fd_d2f = FDHessian(f, 10**-3)

x, iter = newton(
    f, fd_df, fd_d2f,
    np.array([0.]), 
    10**-9,
    100
)

print(f'Solution: {x}')
print(f'Number of iterations: {iter}')
print(f"Derivative f'(x): {fd_df(x)}")
print(f"Second derivative f''(x): {fd_d2f(x)[0][0]}")
```

##### Example #1.3

$f(x,y) = (1-x)^2 + 100(y-x^2)^2$

$\nabla f(x,y) = \begin{bmatrix}
2(x - 1) -400x(y - x^2) \\
200(y - x^2) \\
\end{bmatrix}$

$\nabla^2 f(x,y) = \begin{bmatrix}
2 - 400y + 1200x^2 & -400x \\
-400x & 200 \\
\end{bmatrix}$

```python
def f(x: np.ndarray[np.double]) -> np.double:
    return (1. - x[0])**2 + 100 * (x[1] - x[0]**2)**2

x = np.linspace(0, 2, 10)
y = np.linspace(0, 2, 10)

X, Y = np.meshgrid(x, y)

Z = f([X, Y])

ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("$f(x,y) = (1-x)^2 + 100(y-x^2)^2$")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("f(x, y)")

plt.show()
```

```python
def df(x: np.ndarray[np.double]) -> np.ndarray[np.double]:
    return np.array([
        -2*(1 - x[0]) - 400*x[0]*(x[1] - x[0]**2),
        200*(x[1] - x[0]**2)
    ])

def d2f(x: np.ndarray[np.double]) -> np.ndarray[np.ndarray[np.double]]:
    return np.array([
        [2 - 400*x[1] + 1200*x[0]**2, -400*x[0]],
        [-400*x[0], 200]
    ])

fd_df = FDGradient(f, 10**-3)

error_x = [abs(df([t, 0.])[0] - fd_df([t, 0.])[0]) for t in x]
error_y = [abs(df([t, 0.])[1] - fd_df([t, 0.])[1]) for t in x]

plt.xlabel('x')
plt.grid(True)
plt.plot(x, error_x, label='error_x')
plt.plot(y, error_y, label='error_y')
plt.title('Error of fd df(x) with $\\epsilon = 10^{-3}$')
plt.xlabel("x")
plt.legend()
plt.show()
```

```python
fd_df = FDGradient(f, 10**-4)

error_x = [abs(df([t, 0.])[0] - fd_df([t, 0.])[0]) for t in x]
error_y = [abs(df([t, 0.])[1] - fd_df([t, 0.])[1]) for t in x]

plt.xlabel('x')
plt.grid(True)
plt.plot(x, error_x, label='error_x')
plt.plot(y, error_y, label='error_y')
plt.title('Error of fd df(x) with $\\epsilon = 10^{-4}$')
plt.xlabel("x")
plt.legend()
plt.show()
```

### 2. Automatic Differentiation using JAX library

##### Example #2.1

$f(x) = (x - 2)^2$

$f'(x) = 2(x-2)$

$f''(x) = 2 > 0, \forall x$

```python
import numpy as np
import jax.numpy as jnp
from jax import grad, hessian
from matplotlib import pyplot as plt

def f(x: np.ndarray[np.double]) -> np.double:
    return (x[0] - 2.)**2

def df(x: np.ndarray[np.double]) -> np.ndarray[np.double]:
    return np.array([2*(x[0] - 2)])

def d2f(x: np.ndarray[np.double]) -> np.ndarray[np.ndarray[np.double]]:
    return [[2]]

x = np.linspace(0, 4, 20)

jax_df = grad(f)
error_df = [abs(df([t])[0] - jax_df([t])[0]) for t in x]

jax_d2f = hessian(f)
error_d2f = [abs(d2f([t])[0][0] - jax_d2f([t])[0][0]) for t in x]

plt.xlabel('x')
plt.grid(True)
plt.plot(x, error_df, label="Error of f'(x)")
plt.plot(x, error_d2f, label="Error of f''(x)")
plt.title('Errors')
plt.xlabel("x")
plt.legend(loc='upper center')
plt.show()
```

```python
from modules.optimize.unconstrained.line_search.newton import newton

x, iter = newton(
    f, jax_df, jax_d2f,
    np.array([0.]), 
    10**-9,
    100
)

print(f'Solution: {x}')
print(f'Number of iterations: {iter}')
print(f"Derivative f'(x): {jax_df(x)}")
print(f"Second derivative f''(x): {jax_d2f(x)[0][0]}")
```

```python
jax_df = grad(f)
jax_d2f = hessian(f)

x, iter = newton_eigenvalue_modification(
    f, jax_df, jax_d2f,
    np.array([0., 0.]), 
    10**-9,
    100000
)

print(f'Solution: {x}')
print(f'Number of iterations: {iter}')
print(f'Gradient of f(x, y): {jax_df(x)}')

H = jax_d2f(x)

lam, _ = np.linalg.eig(H)

print(f'Eigenvalues of Hessian f(x,y): {lam}')
```
